### Планы на проект:

**1) qr vs plu vs svd для полноранговых матриц A $n \times m : n \geq m, rk(A) = m$**

**❌ Гипотеза 1:** qr лучше plu по устойчивости, т.к. у $A^*A$ квадрат обусловленности. Задача - найти примеры, где ломается plu, а qr работает.

**❌ Гипотеза 2:** qr лучше svd по устойчивости на матрицах с маленькими сингулярными значениями (т.к. нам нужно будет найти $\Sigma^{-1}$). Задача - найти примеры, где ломается svd, а qr работает.

**Итог:** ожидается, что первенство по устойчивости займет qr. По времени самым быстрым будет plu, самым медленным svd. По памяти лучший plu, худший svd. 

**Формулы:**

$
A^{\dagger} = ((QR)^*(QR))^{-1}(QR)^* = (R^*Q^*QR)^{-1}R^*Q^* = R^{-1}Q^*.
$

$
A^{\dagger} = (PLU)^{-1}A^*
$ - вычисляется неявно в алгоритме, раскрывать не нужно

$
A^{\dagger} = (V \Sigma^2 V^*)^{-1} V \Sigma U^* = V(\Sigma^2)^{-1} V^* V \Sigma U^* = V (\Sigma^2)^{-1} \Sigma U^* = V \Sigma^{-1} U^*
$

**Примечание:** разложение холецкого нет смысла рассматривать, поскольку оно ничем не отличается от plu подхода, кроме лишнего вычисления $A^*A = R R^*$.
